```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
subhalo_params = np.load('subhalos_params_1e-6.npy')
```

```python
m0 = subhalo_params[:,4]
w  = subhalo_params[:,-1]
```

```python
Nsh,lnm_edges = np.histogram(np.log(m0),bins=100,weights=w)
lnm = (lnm_edges[1:]+lnm_edges[:-1])/2.
m = np.exp(lnm)
dlnm = lnm_edges[1:]-lnm_edges[:-1]
dNshdm = Nsh/dlnm/m
```

```python
plt.loglog(m,m**2*dNshdm)
plt.xlim(1e-7,1e12)
plt.ylim(1e8,1e11)
plt.xlabel(r'$m$ [$M_\odot$]')
plt.ylabel(r'$m^{2}dN_{\rm sh}/dm$ [$M_\odot$]')

plt.show()
```
